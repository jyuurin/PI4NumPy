```python
import numpy as np
```

```python
# ler arquivos de altura e idade
arquivoAlturas = np.loadtxt('alturas.txt')
arquivoAnos = np.loadtxt('anos.txt')
```

```python
# atribuir valores dos txts para uma variavel
arrayAlturas = np.array(arquivoAlturas, dtype=np.float16)
arrayAnos = np.array(arquivoAnos, dtype=int)
```

```python
# testes
print(arrayAlturas[3], arrayAlturas[32], arrayAlturas[50])
print(arrayAnos[3], arrayAnos[32], arrayAnos[50])
```

```python
# Filtro de dados:

# Filtro de dados que satisfaz a logica do exercicio:  
filtrar = np.where((arrayAnos >= 1998) & (arrayAnos <= 2005))

# Transformar em um array
arrayIndiceFiltrado = np.array(filtrar, dtype=int)

# Tamanho do array
tamanhoArray = arrayIndiceFiltrado.size 

# Anos com base nos indices
arrayAnosFiltrado = arrayAnos[arrayIndiceFiltrado]

# Alturas com base nos indices
arrayAlturasFiltrado = arrayAlturas[arrayIndiceFiltrado]

# Soma de alturas
SomaAlturas = arrayAlturasFiltrado.sum()
```

```python
print("DADOS ENCONTRADOS:\n")
print("\t- Numero de dados filtrados/Tamanho do Array: ", tamanhoArray)
print("\t- A soma das alturas: ", SomaAlturas)
print("\t- Media de alturas final: ", SomaAlturas / tamanhoArray)
print("\t- Media de alturas arredondada: ", round(SomaAlturas / tamanhoArray, 3))
print("__________________________________________________________________________")

print("Visualização de arrays filtrados: \n")

print("Array de indices filtrados: \n", arrayIndiceFiltrado)
print("__________________________________________________________________________")
print("Array de anos filtrados: \n", arrayAnosFiltrado)
print("__________________________________________________________________________")
print("Array de alturas filtradas: \n", arrayAlturasFiltrado)
print("__________________________________________________________________________")
```
